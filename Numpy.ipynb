```python
#taking input of arrays and creating matrix
import numpy as np
rows =int(input())
cols = int(input())
matrix = []
print("Enter elements row by row:")
for i in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)
A = np.array(matrix)
print("Matrix A:")
print(A)

```

```python
#matrix Addition
print(A + A)
```

```python
#matrix Subtraction
print(A-A)
```

```python
#matrix Multiplication
a=np.dot(A,A)
print(a)
```

```python
#matrix determinant
print(np.linalg.det(A))
```

```python
#Matrix Transpose
print(A.T)
```
