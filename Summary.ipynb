```python
from numpy import var
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\manth\Downloads\record.csv")
df = df.drop_duplicates()
means=df["Weekly_Sales"].mean()
medians=df["Weekly_Sales"].median()
modes=df["Weekly_Sales"].mode()[0]
stds=df["Weekly_Sales"].std()
vars=df["Weekly_Sales"].var()
print("standard deviation=",stds)
print("mean=",means)
print("median=",medians)
print("mode=",modes)
print("variance=",vars)
```
