```python
import pandas as pd
df=pd.read_csv(r"C:\Users\manth\Downloads\record.csv")
print(df.info())
print(df.describe())  
print(df.isnull().sum()) 
```

```python
df = df.drop_duplicates()
df["Weekly_Sales"] = df["Weekly_Sales"].fillna(df["Weekly_Sales"].mean)
df["Date"] = df["Date"].str.strip()
```

```python
summary = df.groupby("Date")["Weekly_Sales"].agg(["mean", "max", "count"])
print(summary)
```
