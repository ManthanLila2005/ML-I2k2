```python
#simulation of a coin toss
import numpy as np
n=int(input("no. of toss="))
h=0
t=0
random= np.random.randint(0, 2, n)
for i in range(0,n):
    if random[i]==0:
        h=h+1
    else:
        t=t+1
print ("heads=",h)
print("tails=",t)
print("P(heads)=",h/n)
print("P(tails)=",t/n)
```

```python
#simulation of a die rolling
import numpy as np
n=int(input("no. of rolls="))
a=[0]*6
random= np.random.randint(1, 7, n)
for i in range(0,6):
    for j in range(0,n):
        if i+1==random[j]:
            a[i]=a[i]+1
for i in range(0,6):
    print(i+1,"=",a[i])
c=int(input("enter the no you want probabilty of"))
print(a[c-1]/n)
```

```python
# conditional probablity
import numpy as np
a=int(input("red="))
b=int(input("blue="))
c=int(input("green="))
d=[a,b,c]
e=[0]*3
h=int(input("the color the ball was of red=0, blue=1, green=2"))
for i in range(0,3):
    if h==i:
        d[i]=d[i]-1
f=[0]*(a+b+c-1)
n=int(input("no. of simulation="))
random= np.random.randint(1,a+b+c, n)
for i in range(0,a+b+c-1):
    for j in range(0,n):
        if i+1==random[j]:
            f[i]=f[i]+1
for i in range(0,3):
    for j in range(0,d[i]):
        e[i]=e[i]+f[j]
print("P(red)=",e[0]/n)
print("P(blue)=",e[1]/n)
print("P(green)=",e[2]/n)
```
