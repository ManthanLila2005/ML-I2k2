```python
#simulation of a coin toss
import numpy as np
n=int(input("no. of toss="))
h=0
t=0
random= np.random.randint(0, 2, n)
for i in range(0,n):
    if random[i]==0:
        h=h+1
    else:
        t=t+1
print ("heads=",h)
print("tails=",t)
print("P(heads)=",h/n)
print("P(tails)=",t/n)
```

```python
# conditional probablity
import numpy as np
a=int(input("red="))
b=int(input("blue="))
c=int(input("green="))
d=[a,b,c]
e=[0]*3
h=int(input("the color the ball was of red=0, blue=1, green=2"))
for i in range(0,3):
    if h==i:
        d[i]=d[i]-1
f=[0]*(a+b+c-1)
n=int(input("no. of simulation="))
random= np.random.randint(1,a+b+c, n)
for i in range(0,a+b+c-1):
    for j in range(0,n):
        if i+1==random[j]:
            f[i]=f[i]+1
for i in range(0,3):
    for j in range(0,d[i]):
        e[i]=e[i]+f[j]
print("P(red)=",e[0]/n)
print("P(blue)=",e[1]/n)
print("P(green)=",e[2]/n)
```

```python
#simulation of 2 coin toss
import numpy as np
n=int(input("no. of toss="))
r1= np.random.randint(0, 2, n)
r2= np.random.randint(0, 2, n)
c1 = 0
c2 =0
c3=0
for i in range(0,n):
    if r1[i]==r2[i]:
        c1=c1+1
        if r1[i]==1:
            c2=c2+1
        else:
            c3=c3+1
print("P(both same)",c1/n)
print("P(both heads)",c2/n)
print("P(both tails)",c3/n)
print("P(both Different)",(n-c1)/n)
```

```python
#predicting weather when sunny next sunny 80%
import random
n =int(input("no. of days"))
state = "S" 
c= 0
for _ in range(n):
    if state == "S":
        if (random.random() < 0.8):
            state = "S" 
        else:
            state="R"
    else:
        if (random.random() < 0.6):
            state = "R" 
        else:
            state="S"

    if state == "S":
        c += 1
print("P(Sunny) =", c / n)
print("P(Rainy) =", (n-c) / n)
```
