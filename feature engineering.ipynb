```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\manth\Documents\PK.csv")
print(df.info())
```

```python
df["date of birth"]=pd.to_datetime(df["date of birth"],dayfirst="true")
print(df.info())
```

```python
df["year_born"]=df["date of birth"].dt.year
current=int(2026)
df["age"]=current-df["year_born"]
print(df)
```

```python
df["full Name"]=df["last Name"]+" "+df["first name"]
df["income level"]=df["salary"].apply(
    lambda x:"high" if x>1000000000 else "low"
)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
df["level"]=a.fit_transform(df["income level"])
```
