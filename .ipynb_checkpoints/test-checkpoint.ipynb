```python
#Palindrome no.
a=input()
b=list(a)
c=b
count=0
for i in range(0,len(b)):
    if(c[len(b)-(i+1)]==b[i]):
        count=count+1
if (count==len(b)):
    print("palindrome")
else:
    print("not palindrome")
```

```python
a=int(input())
count=0
b=a
c=0
while a>=1:
    a=a/10
    count=count+1
a=b
for i in range(0,count):
    d=b%10
    e=d
    b=int(b/10)
    for i in range(1,count):
        e=d*e
    c=c+e
if(c==a):
    print("armstrong")
else:
    print("not armstrong")
```

```python
# factorial using a function
def fact(n):
    if n<=1:
        return 1
    else:
        return n*fact(n-1)
n=int(input())
print(fact(n))
```

```python
# prime no using a function
def prime(n):
    count=0
    if n<=1:
        return "not prime"
    else:
        for i in range(2,n):
            if n%i!=0:
                count=count+1
    if count==(n-2):
        return("prime")
    else:
        return "not prime"
n=int(input())
print(prime(n))
```

```python
# GCD
l=int(input())
v=int(input())
c=0
for i in range(1,l+1):
    if (l%i==0):
        if(v%i==0):
            c=i
print(c)
```

```python
#lcm
l=int(input())
v=int(input())
a=0
b=0
c=l*v
for i in range(1,l+1):
    a=v*i
    for j in range(1,v+1):
        b=l*j
        if a==b:
            if c>b:
                c=b
print(c)
```

```python
#strong no.
def fact(n):
    if n<=1:
        return 1
    else:
        return n*fact(n-1)
a=int(input())
d=a
c=0
for i in range(0,10):
    if a>=1:
        b=a%10
        a=int(a/10)
        c=c+fact(b)
if d==c:
    print("strong no.")
else:
    print("not strong")
```

```python
#pattern program
a=int(input())
for i in range(1,a+1):
    print(" "*(a-i),end="")
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide():
    return a/b
a=int(input())
b=int(input())
c=input("operation: -,+,*,/")
if c=="-":
    print(subtract(a,b))
if c=="+":
    print(add(a,b))
if c=="/":
    print(divide(a,b))
if c=="*":
    print(multiply(a,b))
```
