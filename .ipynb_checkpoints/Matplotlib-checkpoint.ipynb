```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\manth\Downloads\record.csv")
df = df.drop_duplicates()
df["Weekly_Sales"] = df["Weekly_Sales"].fillna(df["Weekly_Sales"].mean)
new=df.groupby("Date")["Weekly_Sales"].mean()
print(new)
```

```python
plt.plot(new.index,new.values)
plt.xlabel("Date")
plt.ylabel("Weekly_Sales")
plt.title("sales record")
plt.show()
```

```python
plt.bar(new.index,new.values)
plt.xlabel("Date")
plt.ylabel("Weekly_Sales")
plt.title("sales record")
plt.show()
```

```python
plt.barh(new.index, new.values)
plt.xlabel("Weekly_Sales")
plt.ylabel("Date")
plt.title("sales record")
plt.show()
```

```python
plt.scatter(new.index,new.values)
plt.xlabel("Date")
plt.ylabel("Weekly_Sales")
plt.title("sales record")
plt.show()
```
