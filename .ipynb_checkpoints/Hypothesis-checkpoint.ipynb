```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
sample_data1=[23,30,28,27,29,27,21,26,28,22]
sample_data2=[22,28,26,25,27,25,20,23,26,20]
alpha=0.05
```

```python
x,y=stats.ttest_ind(sample_data2,sample_data1)
print ("t_stat=",x)
print ("P_value=",y)
```

```python
if y < alpha:
    print("Reject Null Hypothesis")
    print("The two groups are significantly different")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant difference between groups")
```

```python
std=np.std(sample_data1)
mean=np.mean(sample_data1)
```

```python
n=len(sample_data1)
t_critical = stats.t.ppf((1 + (1-alpha)) / 2, df=n-1)
print(t_critical)
```

```python
margin = t_critical * (std / np.sqrt(n))
print(margin)
```

```python
print("95% Confidence Interval:")
print(mean - margin, mean + margin)
```
