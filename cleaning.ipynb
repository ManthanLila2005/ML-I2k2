```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv(r"D:\games\Book2.csv")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
df = df.dropna(subset=["innings"])
df["4s"]=df["4s"].fillna(df["4s"].mean())
df=df.drop_duplicates()
```

```python
print(df.dtypes)
```

```python
df["date"]=pd.to_datetime(df["date"])
print(df.dtypes)
```

```python
Q1 = df["matches"].quantile(0.25)
Q3 = df["matches"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[(df["matches"] >= lower) &
        (df["matches"] <= upper)]
print(df.head())
```

```python
df["player name"]=df["player name"].str.lower()
df["player name"]=df["player name"].str.strip()
print(df["player name"])
```

```python
a=MinMaxScaler()
df["matches"]=a.fit_transform(df[["matches"]])
print(df)
```

```python
b=StandardScaler()
df["matches"]=b.fit_transform(df[["matches"]])
print(df)
```
