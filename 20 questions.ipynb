```python
#3 Check if two strings are anagrams
a=input()
b=input()
c=list(a)
d=list(b)
count=0
for i in range(0,len(c)):
    for j in range(0,len(d)):
        if c[i]==d[j]:
            count = count+1
            break
if count==len(c):
    print("true")
else:
    print("false")
            
```

```python
#5 Rotate matrix by 90 degrees
a,b=int(input()),int(input())
arr = [[0]*a for _ in range(b)]
brr = [[0]*a for _ in range(b)]
for i in range(0,a):
    for j in range(0,b):
        arr[i][j]=int(input())
for i in range(0,a):
    for j in range(0,b):
        brr[j][(a-1)-i]=arr[i][j]
print(arr)
print(brr)
```

```python
#8 write a function to find factorial
def fact(n):
    if n==0:
        return 1
    else:
        return n*(fact(n-1))
n=int(input())
print(fact(n))
```

```python
#9 print a triangle pattern
n=int(input("enter the number of elements"))
a=n
for i in range(1,n*2):
  if i%2==1:
      print(" "*(a),end="")
      print("*"*i)
      a=a-1
```

```python
# 10 find the index of elements in a list whose sum is equal to target
nums=list(map(int,input().split()))
target= int(input())
for i in range(0,len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i]+nums[j]==target:
            arr=[0]*2
            arr[0]=i
            arr[1]=j
            print(arr)
```

```python
#11 convert roman to numbers
s=input()
a=list(s)
total=0
for i in range(0,len(a)):
    if a[i]=="I":
        a[i]=1
    if a[i]=="V":
        a[i]=5
    if a[i]=="X":
        a[i]=10
    if a[i]=="L":
        a[i]=50
    if a[i]=="C":
        a[i]=100
    if a[i]=="D":
        a[i]=500
    if a[i]=="M":
        a[i]=1000
for i in range(0,len(a)-1):
    if a[i]<a[i+1]:
        total=total-a[i]
    else:
        total=total+a[i]
total=total+a[len(a)-1]
print(total)
```
